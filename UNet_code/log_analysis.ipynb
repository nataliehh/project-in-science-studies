```python
import os

base_path = './networks/'
model = '2024_04_16-16_37_36_encoder_efficientnet-b0_weights_imagenet_epochs_100_batchsize_16_lr_0.001'

model_path = os.path.join(base_path, model)
log_path = os.path.join(model_path, 'log.txt')

with open(log_path, 'r') as f:
    content = f.readlines()
len(content)
```

```python
import re
import numpy as np
train_avgs = []
val_avgs = []
for line in content:
    if 'Train_loss' in line or 'Valid_loss' in line:
        elements = re.split(r'[\t ]', line)
        idx = elements.index('Avg')
        loss = elements[idx+1]
        try:
            avg = float(loss)
        except:
            print(line)
            avg = None
        if 'Train_loss' in line:
            train_avgs.append(avg)
        if 'Valid_loss' in line:
            val_avgs.append(avg)
```

```python
import matplotlib.pyplot as plt
plt.plot(train_avgs)
plt.plot(val_avgs)
plt.show()
```
